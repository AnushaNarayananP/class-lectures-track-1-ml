```python
class programming_assignment_3_Module_1:
  def one_tailed_test(self,data):
    ztest ,pval = stests.ztest(data, x2=None, value=data.mean())
    print(float(pval))
    print(ztest)
    if pval<0.05:
      print("reject null hypothesis")
    else:
      print("accept null hypothesis")
    if ztest < 1.645 :
      print('Left_tailed_test')
    else :
      print('Right_tailed_test')
  def graph(self,data):
    kde = stats.gaussian_kde(data)
    t_range = np.linspace(-3.5,30,129)
    plt.plot(t_range,kde(t_range))
    p=(data-data.mean())/data.std()
    plt.plot(t_range,p)
    plt.show()
```

```python
c=programming_assignment_3_Module_1()
c.one_tailed_test(df1['Cyclonic Disturbances - TOTAL'])
```
