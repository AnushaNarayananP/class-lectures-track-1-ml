Covariance Matrix

```python
C=np.cov(X_meaned,rowvar=False)
C
```

Eigen Value and Eigen Vector of Covariance Matrix

```python
eval,evec=np.linalg.eig(C)
```

```python
eval
```

```python
Sum_of_eval=np.sum(eval)
Sum_of_eval
```

Number of Information Conserved

```python
percentage_eval=eval*100/Sum_of_eval
percentage_eval
```

```python
sorted_index=np.argsort(eval)[::-1]
sorted_eval=eval[sorted_index]
```

```python
sorted_evec=evec[:,sorted_index]
sorted_evec
```

```python
def loss_of_information_less(percentage_eval):
  total=0
  count=0
  for i in percentage_eval:
    total+=i
    count+=1
    if total >=90:
      break
  return count

```

```python
evec_subset=sorted_evec[:,0:loss_of_information_less(percentage_eval)]
print(evec_subset)
```

Concising the algorithm

```python
class programming_assignment_3_Module2:
  df1 = pd.read_csv(io.BytesIO(uploaded['water.csv']))
  new_data=df1.to_numpy()
  def loss_of_information_less(self,percentage_eval):
    total=0
    count=0
    for i in percentage_eval:
      total+=i
      count+=1
      if total >=90:
        break
    return count
  
  def PCA_without_loss_of_information(self,new_data) :
     X_meaned = new_data -np.mean(new_data,axis=0)
     C=np.cov(X_meaned,rowvar=False)
     eval,evec=np.linalg.eig(C)
     Sum_of_eval=np.sum(eval)
     percentage_eval=eval*100/Sum_of_eval
     sorted_index=np.argsort(eval)[::-1]
     sorted_eval=eval[sorted_index]
     percentage_eval_index=np.argsort(percentage_eval)[::-1]
     evec_subset=sorted_evec[:,0:loss_of_information_less(percentage_eval)]
     X_reduced= (np.dot(evec_subset.T,X_meaned.T)).T
     Data=pd.DataFrame(X_reduced)
     return Data
     
```
