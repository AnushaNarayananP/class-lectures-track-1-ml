```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y=np.array([3,4,2,4,5])
y_pred =[2.8,3.2,3.6,4.0,4.4]
```

```python
plt.scatter(x,y)
plt.plot(x,y_predicted,color ='r',marker = '*')
plt.show()
```

```python
def mse_loss(y_true,y_pred):
    squared_error = (y_true - y_pred)**2
    sum_squared_error = np.sum(squared_error)
    loss = sum_squared_error/y_true.size
    print(loss)
mse_loss(y,y_pred)    
```

```python
plt.scatter(df['Area'],df['Price'])
plt.show()
```

```python
X = df[['Area']]
y = df[['Price']]
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.coef_
```

```python
model.predict([[4500],[2000]])
```

```python
plt.scatter(df['Area'],df['Price'] ,label = 'Scatter plot')
plt.plot(df['Area'],y_pred,marker='*',label = 'regression' ,color='r')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('House Price Prediction')
plt.legend()
plt.show()
```
