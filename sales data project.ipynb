```python
def new_column(a):
    if (a<pd.to_datetime('2012-01'))==True:
        p=1
    elif ((a>=pd.to_datetime("2012-01") )& (a<pd.to_datetime('2013-01')))==True:
        p=2
    elif ((a>=pd.to_datetime("2013-01"))& (a<pd.to_datetime('2014-01')))==True:
        p=3
    else:
        p=4
    return p
```

```python
train_set['year_division']=train_set['Order Date'].apply(new_column)
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for keys,subdf in df.groupby('year_division'):
        m= np.mean(subdf['Sales'])
        s=np.std(subdf['Sales'])
        reduced_df =subdf[(subdf['Sales']>(m-0.2*s))&(subdf['Sales']<(m+0.2*s))]
        df_out = pd.concat([df_out,reduced_df],ignore_index= True)
    return df_out
```

```python
df4 = remove_pps_outliers(train_set)
```

```python
def remove_pps_outliers1(df):
    df_out = pd.DataFrame()
    for keys,subdf in df.groupby('Quantity'):
        m= np.mean(subdf['Shipping Cost'])
        s=np.std(subdf['Shipping Cost'])
        reduced_df =subdf[(subdf['Shipping Cost']>(m-1*s))&(subdf['Shipping Cost']<(m+1*s))]
        df_out = pd.concat([df_out,reduced_df],ignore_index= True)
    return df_out
```

```python
class sales_data_project:
    
    print(df_02.info())
    print(df_02.duplicated().sum(),' duplicates')
    def separate_to_trainset_and_test_set(self,a):
        df1=a.drop(['Profit'],axis=1)
        y = a['Profit']
        X_train,X_test,y_train,y_test = train_test_split(df1,y,test_size=0.20,random_state=42)
        train_set=pd.concat([X_train,y_train],axis=1)
        test_set=pd.concat([X_test,y_test],axis=1)
        train_set.to_csv(r'C:\\Users\\LENOVO\\PG Diploma in Data Science\\train_set.csv')
        test_set.to_csv(r'C:\\Users\\LENOVO\\PG Diploma in Data Science\\test_set.csv')
        return train_set,test_set,y_test,X_test,y_train,X_train
    def understanding_data(self,b):
        b.drop(['Row ID','Order ID','Customer ID','Customer Name','City','State','Postal Code','Product ID'],axis=1,inplace=True)
        b['Ship Date']=pd.to_datetime(b['Ship Date'],infer_datetime_format=True)
        b['Order Date']=pd.to_datetime(b['Order Date'],infer_datetime_format=True)
        b['no.of_days_taken']=b['Ship Date']-b['Order Date']
        b['no.of_days_taken']=b['no.of_days_taken'].apply(lambda x: int(str(x).split(' days',maxsplit=1)[0]))
        plt.style.use('seaborn')
        b.hist(bins=50,figsize=(20,15),ec='b')
        plt.show()
        return b
    def new_column(self,a):
        if (a<pd.to_datetime('2012-01'))==True:
            p=1
        elif ((a>=pd.to_datetime("2012-01") )& (a<pd.to_datetime('2013-01')))==True:
            p=2
        elif ((a>=pd.to_datetime("2013-01"))& (a<pd.to_datetime('2014-01')))==True:
            p=3
        else:
            p=4
        return p
    def separating_yearwise(self,b):
        b['year_division']=b['Order Date'].apply(new_column)
        return b
    def remove_pps_outliers(self,df):
        df_out = pd.DataFrame()
        for keys,subdf in df.groupby('year_division'):
            m= np.mean(subdf['Sales'])
            s=np.std(subdf['Sales'])
            reduced_df =subdf[(subdf['Sales']>(m-0.2*s))&(subdf['Sales']<(m+0.2*s))]
            df_out = pd.concat([df_out,reduced_df],ignore_index= True)
        return df_out 
    def remove_pps_outliers1(self,df):
        df_out = pd.DataFrame()
        for keys,subdf in df.groupby('Quantity'):
            m= np.mean(subdf['Shipping Cost'])
            s=np.std(subdf['Shipping Cost'])
            reduced_df =subdf[(subdf['Shipping Cost']>(m-1*s))&(subdf['Shipping Cost']<(m+1*s))]
            df_out = pd.concat([df_out,reduced_df],ignore_index= True)
        return df_out
    def model(self,c):
        c.drop(['no.of_days_taken','year_division','Country','Order Date','Ship Date','Product Name'],axis=1,inplace=True)
        df5=pd.get_dummies(data=c,columns=['Ship Mode','Segment','Category','Sub-Category','Market','Region','Order Priority'],drop_first=True)
        X=df5.drop(['Profit'],axis=1)
        y=df5.Profit
        cross_val_score(LinearRegression(),X,y,cv=4)
        
    def prediction(self,X_test,y_test):
        X_test.drop(['Row ID','Order ID','Customer ID','Customer Name','City','State','Postal Code','Product ID','Country','Order Date','Ship Date','Product Name'],axis=1,inplace=True)
        X_test=pd.get_dummies(data=X_test,columns=['Ship Mode','Segment','Category','Sub-Category','Market','Region','Order Priority'],drop_first=True)
        Y_pred = cross_val_predict(LinearRegression(),X_test,y_test, cv=4)
        y_pred = pd.DataFrame(Y_pred)
        return Y_pred,y_pred
    def final_model_graph(self,y_test,y_pred):
        sns.displot(y_test-Y_pred)
    def predicted_csv(self,y_pred,test_set):
        final_test_set=pd.concat([test_set,y_pred],axis=1)
        final_test_set.to_csv(r'C:\\Users\\LENOVO\\PG Diploma in Data Science\\final_test_set.csv')
```
